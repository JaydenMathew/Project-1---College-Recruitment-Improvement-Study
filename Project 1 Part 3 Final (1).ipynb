#### The following code uses the isin function to sort through df2 (the updated dataframe) and isolate the values in the 'CONTINUED INTEREST (1-5)' that are between 4 and 5 due to the question prompt. Then, the code creates another variable that groups the columns that start with 'INTEREST: ' so that they can be viewed together. The next line utilizes the users who answered with a 4 or 5 and takes the count of their 'Yes' values so that they can be displayed. The rest of the code visualizes the interested counts and labels all the axises. The xticks and yticks functions are used to adjust the display of the graph so that it is more visually appealing.

```python
interested_users = df2[df2['CONTINUED INTEREST (1-5)'].isin([4, 5])] #Saves values in the 'CONTINUED INTEREST' column to the interested_users variable if the value in this column is 4 or 5
interest_columns = [col for col in interested_users.columns if col.startswith('INTEREST: ')] #Groups all the columns that start with 'INTEREST: '
interest_yes_counts = interested_users[interest_columns].apply(lambda x: x.eq('Yes').sum()) #Takes the count of all yes values in all of the columns that begin with 'INTEREST: '

# Following code displays and visualizes the code as done previously
plt.figure(figsize=(10, 8))
interest_yes_counts.plot(kind='bar', color='lightblue') #Color is set and type of graph is set
plt.title('Computing Interests of Non-Majors Who Are Interested in Continuing Computing')
plt.xlabel('Computing Interest') #Sets an x-axis label
plt.ylabel('Count of Positve Responses') #Sets a y-axis label
plt.xticks(rotation=45, ha='right') #Rotates the tick labels on the x-axis so that they can be read
plt.yticks(range(0, 15 + 1, 1)) #Changes the tick marks on the y-axis so that they go from 0 to 15 by intervals of 1
plt.show()
```

## From the data in the graph above, it can be seen that out of all the non-majors who are interested in continuing to learn computing at CCM, the majority (11 students) are interested in cybersecurity. This count is quickly followed by computer programming and hardware installation & repair (both of which consist of 10 students). The lowest count for interest is between web development, app development, data analytics, and game design, of which all have 8 students. Using this, CCM faculty can open more classes for the classes with higher interest counts so that they are more accessible for students who wish to continue computing. 

#### The following code calculates the count of rows in df2 for each combination of 'Age' and 'Current Class', and then it arranges this data in a tabular format where 'Age' becomes the index, 'Current Class' becomes the columns, and the counts are the values. The rest of the code plots this tabular formatted data in a bar styled graph that uses the stacked function so that each bar can have different representations (within the bar) depending on the age range. Other features of the code are the make the graph more visually appealing and set up a legend so that the graph is better understood. 

```python
impact_counts = df2.groupby(['Age', 'Current Class']).size().unstack(fill_value=0) #Counts the rows in df2 that have a combination of 'Age' and 'Current Class' and makes this combination in a tabular format

#The following code follows the same order as the above graphs for formatting and display.
plt.figure(figsize=(8, 10))
impact_counts.T.plot(kind='bar', stacked=True, color=['red', 'purple', 'green','blue','black'])
plt.title("Distribution of Varied Ages in Respondent's Classes")
plt.xlabel('Current Class')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.yticks(range(0, 50 + 1, 5))
plt.legend(title='Age Range', bbox_to_anchor=(1, 1), loc='upper left') #This code creates a legend so that the graph is better understood, and makes the legend box located in an area that does not obstruct the view of the graph
plt.show()
```

## From the bar graph above, it can be analyzed that for all three classes (CMP 101, CMP 126, and CMP 135) the majority of non-majors are between 19-20 years olds, which take about 50% of all classes. In contrast, CMP101 and CMP135 are least populated with 35-64 year olds, while CMP126 is least populated with students 18 and younger. Using this, CCM faculty can understand age ranges for student attending their current computing classes, and can adjust their instruction periods so that they better connect with students (For instance, younger students have shorter attention spans than adults and prefer being more hands on, while adults may take these classes after work and need more relaxed teaching styles).

#### The following code filters df2 to find regular expressions for columns that begin with "FACTOR: " and save these columns into a variable (factor_columns). This is a new method I found that is similar to the col.startwith() that was used in a previous graph. Next, the code takes the count of unique values in these filtered columns and saves this code to impact_counts. Then, impact_counts is transposed, meaning that the unique values are set to the rows of the tabular data, and the original values are held as the columns (making it better to visualize). The plotting code afterwards is used to set colors, style, and display the graph is the most appealing way. 

```python
factor_columns = df2.filter(regex='^FACTOR:') #Filters df2 for all columns that start with 'FACTOR: '
impact_counts = factor_columns.apply(lambda col: col.value_counts()) #Takes the count of all unique values in the factor_column variable
impact_counts = impact_counts.transpose() #Tranposes this data so that it is in tabular format where unique values are the rows and the original values are the columns

#The following code is the same as the previous graphs (any additional code will be noted)
plt.figure(figsize=(12, 8))
impact_counts.plot(kind='bar', stacked=True, color=['red', 'orange', 'green']) #The stacked function makes the data in the same columns stack on top of each other (how the impact levels are stacked)
plt.title('Influential Factors on CCM Attendees')
plt.xlabel('Factors')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() #Displays the graph in a tigher layout so that it is more organized and appealing
plt.legend(title='Impact Level', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

## The above graph shows the count of different impact levels that certain factors had on why students attended CCM. From the graph, it can be seen that the highest impacting factors were cost and location, while the lowest impacting factor was credit in highschool. CCM faculty can use this data to prioritize certain factors of CCM to increase CCM attendees. For instance, they can find ways to decrease cost or increase financial aid so that more students attend.
